```python
import re 
```

```python
with open("/Users/nandy/Downloads/AI/cat-resnet50.txt", "r") as f :
    txt = f.read()    
f.close()
```

```python
#Cat 
```

```python
def strip_(string, txt):
    pattern = f"{string} (\d+.\d+)"
    return [float(x) for x in re.findall(pattern, txt)]
```

```python
resnet50_train_loss = strip_("Training Loss:", txt)
resnet50_test_loss = strip_("Validation Loss:", txt)
resnet50_train_acc = strip_("Training Accuracy:", txt)
resnet50_test_acc = strip_("Validation Accuracy:", txt)
```

```python
with open("/Users/nandy/Downloads/AI/cat-mv2.txt", "r") as f :
    txt = f.read()    
f.close()
```
